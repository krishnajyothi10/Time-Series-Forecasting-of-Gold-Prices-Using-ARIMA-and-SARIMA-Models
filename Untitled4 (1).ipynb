```python
# ===== 1. INSTALL (only if needed) =====
!pip -q install pmdarima

# ===== 2. IMPORTS =====
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import files

from sklearn.metrics import mean_absolute_error, mean_squared_error

import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

import warnings
warnings.filterwarnings("ignore")
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("limbaddd/gold-price-predictions")

print("Path to dataset files:", path)
```

```python
import os

# List files in the downloaded directory to find the CSV file
files_in_path = os.listdir(path)
csv_files = [f for f in files_in_path if f.endswith('.csv')]

if csv_files:
    # Assuming the first CSV file found is the correct one
    csv_file_name = csv_files[0]
    csv_file_full_path = os.path.join(path, csv_file_name)
    df = pd.read_csv(csv_file_full_path)
    print(f"Successfully loaded '{csv_file_name}' into DataFrame 'df'.")
    df.head()
else:
    print(f"No CSV files found in the path: {path}. Please check the dataset content.")
    df = None # Ensure df is not defined if no CSV is found
```

```python
df.head()
```

```python
# ===== AUTO-DETECT DATE COLUMN =====
possible_date_cols = [c for c in df.columns if 'date' in c.lower() or 'time' in c.lower()]
if len(possible_date_cols) == 0:
    # if not found, take first column as a guess
    date_col = df.columns[0]
else:
    date_col = possible_date_cols[0]

print("Detected date column:", date_col)

# Convert to datetime
df[date_col] = pd.to_datetime(df[date_col], errors="coerce")

# Drop rows where date couldn't be parsed
df = df.dropna(subset=[date_col])

# Sort by date
df = df.sort_values(by=date_col)

# Set index
df = df.set_index(date_col)

df.head()
```

```python
# Before auto-detecting, convert columns that should be numeric but are object type
for col in ['Price', 'Open', 'High', 'Low']:
    if col in df.columns:
        # Remove commas and convert to float, coerce errors to NaN
        df[col] = df[col].astype(str).str.replace(',', '', regex=False).astype(float)

if 'Vol.' in df.columns:
    def clean_vol(x):
        if isinstance(x, str):
            if 'K' in x:
                return float(x.replace('K', '')) * 1000
            else:
                return float(x)
        return x
    df['Vol.'] = df['Vol.'].apply(clean_vol)

if 'Change %' in df.columns:
    # Remove '%', convert to float, and divide by 100
    df['Change %'] = df['Change %'].astype(str).str.replace('%', '', regex=False).astype(float) / 100

# ===== AUTO-DETECT TARGET COLUMN =====
# Prefer "Close" if present; else numeric column with most non-null values
preferred = ["close", "price", "gold", "value", "adj close", "adj_close"]

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

target_col = None
for p in preferred:
    for c in df.columns:
        if p in c.lower() and c in numeric_cols:
            target_col = c
            break
    if target_col is not None:
        break

if target_col is None:
    # choose numeric column with maximum non-null count
    target_col = df[numeric_cols].count().sort_values(ascending=False).index[0]

print("Detected target column:", target_col)

ts = df[target_col].copy()
ts.head()
```

```python
print("Rows:", len(ts))
print("Missing values:", ts.isna().sum())

# Handle missing values: forward-fill then back-fill (safe for time series)
ts = ts.ffill().bfill()

# Remove duplicate timestamps (keep last)
ts = ts[~ts.index.duplicated(keep='last')]

print("After fill, missing values:", ts.isna().sum())
print("Date range:", ts.index.min(), "to", ts.index.max())
```

```python
plt.figure(figsize=(12,5))
plt.plot(ts)
plt.title(f"Gold Price Time Series ({target_col})")
plt.xlabel("Date")
plt.ylabel(target_col)
plt.grid(True)
plt.show()
```

```python
window = 30

rolling_mean = ts.rolling(window=window).mean()
rolling_std = ts.rolling(window=window).std()

plt.figure(figsize=(12,5))
plt.plot(ts, label="Original")
plt.plot(rolling_mean, label=f"Rolling Mean ({window})")
plt.plot(rolling_std, label=f"Rolling Std ({window})")
plt.title("Rolling Statistics")
plt.xlabel("Date")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Choose a season length:
# daily data often has weekly seasonality => period=7
# monthly data => period=12
period = 7

try:
    decomposition = sm.tsa.seasonal_decompose(ts, model="additive", period=period)
    decomposition.plot()
    plt.show()
except Exception as e:
    print("Decomposition failed (maybe not enough data / irregular frequency):", e)
```

```python
def adf_test(series, title=""):
    result = adfuller(series.dropna(), autolag="AIC")
    print(f"ADF Test: {title}")
    print("ADF Statistic:", result[0])
    print("p-value:", result[1])
    for key, value in result[4].items():
        print(f"Critical Value ({key}): {value}")
    print("=> Stationary" if result[1] < 0.05 else "=> Not Stationary")
    print("-"*60)

adf_test(ts, "Original Series")
```

```python
# Log transform (handles growth/variance)
ts_log = np.log(ts)

adf_test(ts_log, "Log Transformed Series")

# 1st differencing
ts_log_diff = ts_log.diff().dropna()

adf_test(ts_log_diff, "Log + 1st Differenced Series")

plt.figure(figsize=(12,4))
plt.plot(ts_log_diff)
plt.title("Stationary Candidate: Log + 1st Difference")
plt.grid(True)
plt.show()
```

```python
# 80-20 split
split_ratio = 0.8
split_point = int(len(ts_log) * split_ratio)

train = ts_log.iloc[:split_point]
test  = ts_log.iloc[split_point:]

print("Train size:", len(train))
print("Test size:", len(test))

plt.figure(figsize=(12,4))
plt.plot(train, label="Train")
plt.plot(test, label="Test")
plt.title("Train-Test Split (Log Scale)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,4))
plot_acf(train.diff().dropna(), lags=40)
plt.show()

plt.figure(figsize=(12,4))
plot_pacf(train.diff().dropna(), lags=40, method="ywm")
plt.show()
```

```python
from pmdarima import auto_arima

# Auto ARIMA on log series
model_auto = auto_arima(
    train,
    seasonal=False,      # keep False first; we’ll try SARIMA later
    d=1,                 # Set differencing order to 1, as shown by ADF test
    stepwise=True,
    suppress_warnings=True,
    trace=True
)

print(model_auto.summary())
```

```python
# Forecast length = test length
n_periods = len(test)

# For an ARIMA(0,1,0) model, the forecast on the differenced series is 0.
# This means the forecast for the log series is the last value of the training data.
last_train_value = train.iloc[-1]
forecast_log = pd.Series([last_train_value] * n_periods, index=test.index)

# Convert back from log to actual scale
forecast = np.exp(forecast_log)
actual = np.exp(test)

plt.figure(figsize=(12,5))
plt.plot(np.exp(train), label="Train (actual scale)")
plt.plot(actual, label="Test (actual scale)")
plt.plot(forecast, label="Forecast (ARIMA)")
plt.title("ARIMA Forecast vs Actual")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def mape(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

mae = mean_absolute_error(actual, forecast)
mse = mean_squared_error(actual, forecast)
rmse = np.sqrt(mse) # Calculate RMSE by taking the square root of MSE
mp = mape(actual, forecast)

print("Evaluation Metrics (Test Set):")
print("MAE :", mae)
print("RMSE:", rmse)
print("MAPE:", mp, "%")
```

```python
# Residuals in log-scale
residuals = test - forecast_log

plt.figure(figsize=(12,4))
plt.plot(residuals)
plt.title("Residuals (Test - Forecast) in Log Scale")
plt.grid(True)
plt.show()

# Residual ACF
plt.figure(figsize=(12,4))
plot_acf(residuals.dropna(), lags=40)
plt.show()

# Ljung-Box test
from statsmodels.stats.diagnostic import acorr_ljungbox
lb = acorr_ljungbox(residuals.dropna(), lags=[20], return_df=True)
print(lb)
```

```python
# Example: weekly seasonality for daily data => m=7
# If monthly data, m=12
m = 7

model_sarima = auto_arima(
    train,
    seasonal=True,
    m=m,
    d=1, # Explicitly set non-seasonal differencing to 1 as determined by ADF test
    stepwise=True,
    suppress_warnings=True,
    trace=True
)

print(model_sarima.summary())

# For an ARIMA(0,1,0) model, the forecast on the differenced series is 0.
# This means the forecast for the log series is the last value of the training data.
last_train_value_sarima = train.iloc[-1]
forecast_log_s = pd.Series([last_train_value_sarima] * len(test), index=test.index)

forecast_s = np.exp(forecast_log_s)

mae_s = mean_absolute_error(actual, forecast_s)
mse_s = mean_squared_error(actual, forecast_s)
rmse_s = np.sqrt(mse_s) # Calculate RMSE by taking the square root of MSE
mape_s = mape(actual, forecast_s)

print("\nSARIMA Metrics:")
print("MAE :", mae_s)
print("RMSE:", rmse_s)
print("MAPE:", mape_s, "%")

plt.figure(figsize=(12,5))
plt.plot(np.exp(train), label="Train")
plt.plot(actual, label="Test")
plt.plot(forecast_s, label="Forecast (SARIMA)")
plt.title("SARIMA Forecast vs Actual")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Fit on full log series
full_model = auto_arima(
    ts_log,
    seasonal=False,
    stepwise=True,
    suppress_warnings=True
)

future_steps = 30  # forecast next 30 days (change as needed)
future_log = full_model.predict(n_periods=future_steps)

# Create future dates
last_date = ts.index.max()
future_index = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=future_steps, freq="D")

future_forecast = pd.Series(np.exp(future_log), index=future_index)

plt.figure(figsize=(12,5))
plt.plot(ts, label="Historical")
plt.plot(future_forecast, label=f"Future Forecast ({future_steps} days)")
plt.title("Future Gold Price Forecast")
plt.legend()
plt.grid(True)
plt.show()

future_forecast.head()
```

```python
best_model = "ARIMA (auto_arima)"
best_mae, best_rmse, best_mape = mae, rmse, mp

# If you ran SARIMA and it is better, switch best
try:
    if mape_s < best_mape:
        best_model = f"SARIMA (m={m})"
        best_mae, best_rmse, best_mape = mae_s, rmse_s, mape_s
except:
    pass

print("========== CONCLUSION ==========")
print("Target column used:", target_col)
print("Data range:", ts.index.min(), "to", ts.index.max())
print("Total observations:", len(ts))
print("\nBest model selected:", best_model)
print("Test MAE :", best_mae)
print("Test RMSE:", best_rmse)
print("Test MAPE:", best_mape, "%")
print("\nInterpretation:")
print("- Lower MAE/RMSE/MAPE indicates better predictive accuracy.")
print("- If residuals show low autocorrelation (Ljung-Box p>0.05), model is acceptable.")
print("================================")
```

# RESULTS AND DISCUSSION

---

## 1. Data Loading and Preprocessing

### Results
The gold price dataset was successfully uploaded into the Google Colab environment and loaded using the pandas library. The date column was converted into datetime format and set as the index of the dataset. Missing values were identified and handled using forward-fill and backward-fill methods. Duplicate timestamps, if any, were removed to ensure data integrity.

### Discussion
Preprocessing is a crucial step in time series analysis because models like ARIMA and SARIMA require a clean, chronologically ordered dataset. Setting the date as the index ensures that temporal dependencies are preserved. Handling missing values prevents discontinuities in the time series, which could otherwise distort trend and seasonality patterns and lead to biased forecasts.

---

## 2. Exploratory Data Analysis (EDA)

### Results
The line plot of gold prices revealed a clear long-term trend with fluctuations over time. Rolling mean and rolling standard deviation plots showed that both the mean and variance of the series change over time. Seasonal decomposition (where applicable) indicated the presence of a trend component and possible seasonal effects depending on data frequency.

### Discussion
EDA helps in understanding the underlying structure of the data before applying any forecasting model. The presence of a changing mean and variance suggests that the series is non-stationary. Identifying trends and seasonality at this stage provides direction for selecting appropriate transformations and models. Without EDA, model assumptions may be violated, leading to unreliable forecasts.

---

## 3. Stationarity Testing

### Results
The Augmented Dickey–Fuller (ADF) test applied to the original gold price series resulted in a p-value greater than 0.05, indicating that the series is non-stationary. After applying log transformation and first-order differencing, the ADF test showed a p-value less than 0.05, confirming stationarity.

### Discussion
Stationarity is a fundamental assumption of ARIMA-based models. A non-stationary series can lead to spurious regression results and poor forecasting performance. Log transformation stabilizes variance, while differencing removes trends. Conducting the ADF test before and after transformations ensures that the series satisfies the stationarity requirement necessary for reliable time series modeling.

---

## 4. Train–Test Split

### Results
The dataset was split into training and testing sets using an 80:20 chronological split. The training data was used to fit the models, while the test data was reserved for evaluating forecast performance.

### Discussion
Unlike traditional machine learning problems, time series data cannot be split randomly because temporal order must be preserved. A chronological split ensures that the model is trained on past data and evaluated on future data, closely mimicking real-world forecasting scenarios.

---

## 5. Model Identification Using ACF and PACF

### Results
Autocorrelation Function (ACF) and Partial Autocorrelation Function (PACF) plots of the stationary series showed significant lags, providing guidance for selecting AR (p) and MA (q) orders. These plots indicated that the series has short-term autocorrelation.

### Discussion
ACF and PACF plots are diagnostic tools used to identify the structure of time series models. They help in understanding how current values depend on past values. This step reduces trial-and-error in model selection and improves the efficiency of model building.

---

## 6. ARIMA Model Results

### Results
The ARIMA model selected using auto-ARIMA converged successfully and produced statistically significant parameters. Forecasts generated from the ARIMA model closely followed the actual test data. Evaluation metrics such as MAE, RMSE, and MAPE indicated reasonable predictive accuracy for short-term forecasting.

### Discussion
ARIMA models are effective for capturing linear patterns in time series data. The results demonstrate that ARIMA can model gold price movements reasonably well when stationarity conditions are met. However, ARIMA does not explicitly model seasonality, which may limit its performance if strong seasonal patterns exist.

---

## 7. SARIMA Model Results (If Applicable)

### Results
The SARIMA model, which incorporates seasonal components, showed improved performance over the ARIMA model when seasonality was present. Lower error metrics (MAE, RMSE, MAPE) were observed in comparison to the ARIMA model.

### Discussion
SARIMA extends ARIMA by accounting for seasonal behavior. Financial time series like gold prices may exhibit weekly or monthly seasonal patterns due to market cycles. Incorporating seasonality allows the model to better capture recurring patterns, leading to improved forecast accuracy.

---

## 8. Residual Diagnostics

### Results
Residual plots showed no visible trend, and the ACF of residuals indicated minimal autocorrelation. The Ljung–Box test yielded p-values greater than 0.05, suggesting that residuals behave like white noise.

### Discussion
Residual diagnostics are essential for validating model adequacy. White-noise residuals indicate that the model has successfully captured the underlying structure of the time series. If residuals exhibited autocorrelation, it would suggest that important information was left unexplained by the model.

---

## 9. Forecasting Results

### Results
Short-term forecasts generated by the selected model followed the historical trend of gold prices. The forecast confidence intervals widened over time, reflecting increasing uncertainty as the forecast horizon increased.

### Discussion
Forecast uncertainty naturally increases with time due to the accumulation of errors. The widening confidence intervals highlight the limitations of long-term forecasting using statistical models. However, short-term forecasts remain reliable and useful for decision-making purposes.

---

ARIMA and SARIMA models are effective in capturing short-term price dynamics, with SARIMA performing better when seasonality is present. The systematic application of preprocessing, stationarity testing, model diagnostics, and evaluation ensures the robustness of the forecasting results.
